## 1. SCALING <a class="anchor" id="1"> 
[Table of Contents](#0)
- operation increases/reduces(resize) size of an image.
- In Scaling we discussed these categories
  - Preserve Aspect Ratio (height to width ratio of image is preserved)
  - Do not preserve Aspect Ratio, resize only width or height

### 1.1 Preserve Aspect Ratio <a class="anchor" id="1.1"> 
    

#### 1.1.1 Downscaling <a class="anchor" id="1.1.1"> 
[Table of Contents](#0)

```python
import cv2
 
img = cv2.imread('Mapengo_emoji.jpg', cv2.IMREAD_UNCHANGED)
 
print('Original Dimensions : ',img.shape)
 
scale_percent = 60 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<p> The left side of the histogram white indicates the darker shades of intensity while the right side are 
the brighter sides of the pixel values. 

<p> On the x-axis of the histogram we have the pixels (bins) while on the y-axis we have the frequency of the occurence of the pixels.

### Histogram Equalization <a class="anchor" id="B2"> 

[Table of Contents](#0)

<p> Histogram Equalization is a technique for adjusting image intensities to enhance contrast. 
    
<p> It accomplishes this by effectively spreading out the most frequent intensity values i.e the intensity range of the image. 
    
<p> It usually increases the global contrast of images when its usable data is represented by close contrast values, allowing areas of lower local contrast to gain a higher contrast. 

```python
img_hist = cv2.imread("C:/Users/brian.samba/Documents/Computation Intelligence/Computer Vision/Images/archive/data/apple.jpg", 0)
equ = cv2.equalizeHist(img_hist)

plt.imshow(cv2.cvtColor(img_hist, cv2.COLOR_BGR2RGB)), plt.title('original'),plt.show()
plt.imshow(cv2.cvtColor(equ, cv2.COLOR_BGR2RGB)), plt.title('equalized'), plt.show()
```

```python
### Calculate the histogram of the equalized histogram

hist = cv2.calcHist([equ],[0],None,[256], [0,256])
plt.plot(hist), plt.show()

#plt.hist(equ.ravel(),256,[0,256]); plt.show()

```

## Image Transforms <a class="anchor" id="C"> 

[Table of Contents](#0)

<p1> Image transforms are a technique used to convert an image from one domain to another domain i.e from the time/spatial domain to the frequency domain

<p1> Image transfroms do not alter the information of the original image and the operation can be reversed to obtain the original image through the use of inverse transforms

<p1> Image transform operations allow us to simplify calculations and computations i.e operations which involve complex calculations in one domain can be simplified by converting the image to another domain and then transforming it back to the the original domain once the computation has been completed.

<p> Another usefullness of image transfoms is that it enables the extraction of more features from the image, e.g white light can be split using a prism to obtain a spectra of colors, enabling us to know the composition of the light. 

### Fourier Transform <a class="anchor" id="C1"> 

[Table of Contents](#0)

<p> The fourier transform transforms an image into the frequency domain, with the resulting image having a real and an imaginary part
    

```python
img_spatial = cv2.imread("C:/Users/brian.samba/Documents/Computation Intelligence/Computer Vision/Images/archive/data/lena.jpg", 0)

dft = cv2.dft(np.float32(img_spatial), flags=cv2.DFT_COMPLEX_OUTPUT)

# the dft function only accepts float32 and hence the conversion of the image to float32
# the flag part indicates that the function should output an image with an imaginary part

dft_shift = np.fft.fftshift(dft) # shifts the center of the output from the top left
# corner to the center

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))

plt.subplot(121), plt.imshow(img_spatial, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap ='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()

```

####  High Pass Filtering and Performing the Inverse Discrete Fourier Transform <a class="anchor" id="C1.1"> 

[Table of Contents](#0)

```python
## HPF Mask
rows, cols = img_spatial.shape
crow, ccol = int(rows/2), int(cols/2)

mask = np.ones((rows, cols,2), np.uint8)
r = 10
center =[crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = (x - center[0]) ** 2 + (y-center[1]) **2 <= r*r
mask[mask_area] = 0

fshift = dft_shift * mask
fshift_mask_mag = 2000 * np.log(cv2.magnitude(fshift[:,:,0], fshift[:,:,1])+1)

f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0], img_back[:,:,1])

fig = plt.figure(figsize =(12, 12))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_spatial, cmap='gray')
ax1.title.set_text('Input Image')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(magnitude_spectrum, cmap='gray')
ax2.title.set_text('FFT of image')
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(fshift_mask_mag, cmap='gray')
ax3.title.set_text('FFT + Mask')
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(img_back, cmap='gray')
ax4.title.set_text('After Inverse FFT')
plt.show()

```
